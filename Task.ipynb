```python
array = data['Type'].unique()

for k in array:
    print(k)

    for i in data:
        sum = 0
        count = 0
        min = 0
        max = 0
        a = []
        if i != 'Type':
            for j in range(len(data[i])):
                if data['Type'][j] == k:
                    a.append(data[i][j])
                    sum += data[i][j]
                    count += 1
                    if data[i][j] < min:
                        min = data[i][j]
                    if data[i][j] > max:
                        max = data[i][j]
            avg = sum / count
            print(i, avg, max, min)
    
```
